```python
import numpy as np
import matplotlib.pyplot as plt

num_items = 100
skewness_values = [0.9]

plt.figure(figsize=(10, 6))

for skew in skewness_values:
    ranks = np.arange(1, num_items + 1)
    frequencies = 1 / np.power(ranks, skew)
    frequencies /= frequencies.sum()  # Normalize
    plt.plot(ranks, frequencies, label=f'skewness={skew}')

plt.xlabel('Rank')
plt.ylabel('Normalized Frequency')
plt.title('Zipf Distribution for Different Skewness Values')
plt.legend()
plt.grid(True)
plt.show()
```
