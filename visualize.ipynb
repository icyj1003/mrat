```python
import numpy as np
import matplotlib.pyplot as plt

num_items = 200
skewness_values = [1.0]

plt.figure(figsize=(10, 6))

for skew in skewness_values:
    ranks = np.arange(1, num_items + 1)
    frequencies = 1 / np.power(ranks, skew)
    frequencies /= frequencies.sum()  # Normalize
    plt.plot(ranks, frequencies, label=f'skewness={skew}')

plt.xlabel('Rank')
plt.ylabel('Normalized Frequency')
plt.title('Zipf Distribution for Different Skewness Values')
plt.legend()
plt.grid(True)
plt.show()
```

```python
num_vehicles = 30
x = np.random.uniform(0, 2000, num_vehicles)
y = [1 for _ in range(num_vehicles)]  # Constant y-value of 1
plt.figure(figsize=(20, 3))
plt.scatter(x, y, alpha=0.5)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot of Random Points')
plt.grid(True)
plt.show()
```
